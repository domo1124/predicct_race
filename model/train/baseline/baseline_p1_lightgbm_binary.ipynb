# LightGBM  
 - 前走のデータを使った予測モデル
 - 3着までを1、それ以外を0とした2値分類

```python
import numpy as np
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split,StratifiedKFold,KFold,train_test_split
from sklearn import metrics
import lightgbm as lgb
```

```python
#データの読み込み
df = pd.read_csv("../../data/baseline_data.csv")
#target
df["target"] = df.pred_tyaku.apply(lambda x : 1 if x < 4 else 0)
df_past1 = df[(df["past_race_num"]==1)]
```

```python
learn_data = df_past1.sample(frac=1,random_state=1)
X          = learn_data.iloc[:,4:-1].values
y          = learn_data.target.values
x_columns  = learn_data.iloc[:,4:-1].columns

X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,stratify=y,random_state=1)
```

```python
#交差検証の設定
num_kfold = 5
kf = KFold(n_splits=num_kfold,shuffle=True,random_state=1)
```

```python
#param
params = {'boosting_type': 'gbdt',
          'objective': 'binary',
          'metric': 'binary_logloss',
          'learning_rate':0.01,
          'num_iterations':1000,
          'verbosity': -1,
          'early_stopping_rounds': 100,
          'num_leaves': 63,
          'max_depth': 7,
          'seed': 1,
            }
```

```python
#Model Learning
model_set = []
evals_result_set = []
for i,(train_index, test_index) in enumerate(kf.split(X_train, y_train)):
    #lightgbmの学習遷移を格納する配列
    evals_result={}
    #学習遷移のdictのKey
    evals_name=["eval_cv{}".format(i),"train_cv{}".format(i)]
    #LightGBM用のデータに変換
    train_data = lgb.Dataset(X_train[train_index], label = y_train[train_index])
    eval_data  = lgb.Dataset(X_train[test_index],  label = y_train[test_index] , reference= train_data)
    #モデルの学習
    model = lgb.train(params, 
                      train_data, 
                      valid_names=evals_name,
                      valid_sets=[eval_data,train_data],
                      evals_result=evals_result,verbose_eval=False)
    model_set.append(model)
    evals_result_set.append(evals_result)
```

```python
#featuer_importance
def featuer_importance_plot(model,columns):
    feature_imp = pd.DataFrame(sorted(zip(model.feature_importance(),columns)), columns=['Value','Feature'])
    plt.figure()
    sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value", ascending=False).iloc[0:20,:])
    plt.title('LightGBM Features')
    plt.tight_layout()
    plt.show()
#run
for model in model_set:
    featuer_importance_plot(model,x_columns)
```

```python
#auc
def auc_plot(model,pred_data,pred_target):
    data = model.predict(pred_data)
    auc = metrics.roc_auc_score(pred_target, data)
    print(auc)

    fpr, tpr, thresholds = metrics.roc_curve(pred_target, data)
    auc = metrics.auc(fpr, tpr)
    plt.plot(fpr, tpr, label='SVC (AUC = %.2f)'%auc)
    plt.legend()
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.title('ROC curve')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.grid(True)
    plt.show()
#run
for model in model_set:
    auc_plot(model,X_test,y_test)
```

```python
#loss
def loss_plot(train_loss,test_loss):
    print("train (epoch:{}/loss:{})" .format( str(len(train_loss)) , str(train_loss[-1])))
    print("test  (epoch:{}/loss:{})" .format( str(len(test_loss)) , str(test_loss[-1])))
    plt.figure()
    plt.plot(np.arange(0,len(train_loss)),train_loss,label="loss for training")
    plt.plot(np.arange(0,len(test_loss)),test_loss,label="loss for validation")
    plt.legend()
    plt.title('model loss')
    plt.xlabel('epoch')
    plt.ylabel('loss')
    plt.legend(loc='upper right')
    plt.show()
#run
for result in evals_result_set:
    train_key,test_key = (result.keys())
    train_loss = list(result[train_key].values())[0]
    test_loss  = list(result[test_key].values())[0]
    loss_plot(train_loss,test_loss)
```

```python
#K-Fold
for model in model_set:
    ave_set = np.zeros((X_test.shape[0],1))
    pred=model.predict(X_test)
    ave_set += pred.reshape((X_test.shape[0],1))
result_pred = ave_set/5
#log loss
loss = metrics.log_loss(y_test, result_pred)
print("log_loss",loss)
#auc
auc = metrics.roc_auc_score(y_test, result_pred)
print("auc",auc)
fpr, tpr, thresholds = metrics.roc_curve(y_test, result_pred)
auc = metrics.auc(fpr, tpr)
plt.plot(fpr, tpr, label='SVC (AUC = %.2f)'%auc)
plt.legend()
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.title('ROC curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.grid(True)
plt.show()
```
